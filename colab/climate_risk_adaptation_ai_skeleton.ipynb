# 5. Skor Kredit Berbasis Risiko Iklim (Banking Relevance)

Bagian ini menghubungkan data iklim dengan praktik perbankan sesuai framework
**TCFD** (*Task Force on Climate-related Financial Disclosures*) dan regulasi
**OJK SEOJK No. 16/2021** tentang Keuangan Berkelanjutan.

#### Mengapa bank perlu climate risk scoring?

- **Physical risk** (banjir, kekeringan, suhu ekstrem) → NPL naik di daerah terdampak
- **Transition risk** (carbon tax, regulasi hijau) → stranded assets sektor fosil
- **Nilai agunan** properti turun akibat sea-level rise atau bencana berulang
